# IMPORTANT NOTE
. Delhi daradevils was named delhi capitals and Deccan chargers were name sunrisers hyderabad

```python
def change_teams(x):
    if x=='Delhi Daredevils':
        x='Delhi Capitals'
    elif x=='Deccan Chargers':
        x=='Sunrisers Hyderabad'
    elif x=='Rising Pune Supergiant' or x=='Pune Warriors':
        x=='Rising Pune Supergiant'
    return x
```

```python
Matches['team1']=Matches['team1'].apply(lambda x:change_teams(x))
Matches['team2']=Matches['team2'].apply(lambda x:change_teams(x))
Matches['toss_winner']=Matches['toss_winner'].apply(lambda x:change_teams(x))
Matches['winner']=Matches['winner'].apply(lambda x:change_teams(x))
```

```python
#Total wins by every team so far 
Matches.winner.value_counts()
```

```python
plt.figure(figsize=(15,5))
match_wins=Matches.winner.value_counts()
Matches.winner.value_counts().plot.bar()
```

```python
#Total MAtches played by teams so far
tot_match=Matches.team1.value_counts()+Matches.team2.value_counts()
tot_match.sort_values(ascending=False)
```

```python
plt.figure(figsize=(15,5))
tot_match.sort_values(ascending=False).plot.bar()
```

```python
#percentage of team wins
plt.figure(figsize=(15,5))
wins_per=(match_wins/tot_match)*100
wins_per.plot.bar()
```

```python
wins_per.sort_values(ascending=False)
```
